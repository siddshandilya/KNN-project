```python
import pandas as pd

import os
```

```python
dt = pd.read_csv(r"C:\Users\Sidha\Desktop\ml files\nba_2013.csv")
dt.head(5)
```

# Checking for null values


```python
dt.isnull().any()
```

# Filling the null values with mean of the column

```python
dt["fg."].fillna(dt["fg."].mean(),inplace=True)
dt["x2p."].fillna(dt["x2p."].mean(),inplace=True)
dt["efg."].fillna(dt["efg."].mean(),inplace=True)
dt["x3p."].fillna(dt["x3p."].mean(),inplace=True)
dt["ft."].fillna(dt["ft."].mean(),inplace=True)
```

```python
dt.isnull().any()
```

# selecting the numeric data columns

```python
numcols=dt.select_dtypes(exclude=['object']).columns
dt=dt[numcols]
```

```python
dt.head(5)
```

```python
dt = dt.drop(['season_end'], axis = 1)
```

```python
dt.head(5)
```

```python
pt=dt['pts']
dt=dt.drop(['pts'], axis = 1)
```

```python
pt.head(5)
```

```python
dt.head(5)
```

# Normalizing values

```python
dt = dt.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
dt.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# we will make predictions using dt

# we will predict pt

x_train, x_test, y_train, y_test = train_test_split(dt, pt, test_size=0.3, random_state=0)
```

# Finding K value

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics

# checking for increasing neighbour values to see which has the highest regression score

for k in range(10):
    k_val = k + 1
    knn = KNeighborsRegressor(n_neighbors = k_val)
    knn.fit(x_train, y_train) 
    y_pred = knn.predict(x_test)
    print (" The regression score is:",format(metrics.r2_score(y_test, y_pred),'.4f'), "for k_value:", k_val)
```

#  we choose k=8 since it has the highest regression score

```python
knn = KNeighborsRegressor(n_neighbors = 8)
knn.fit(x_train, y_train) 
y_pred = knn.predict(x_test)
print ("Mean Squared Error is:", format(metrics.mean_squared_error(y_test, y_pred), '.7f'))
print ("Regression score is:", format(metrics.r2_score(y_test, y_pred),'.4f'))
```
